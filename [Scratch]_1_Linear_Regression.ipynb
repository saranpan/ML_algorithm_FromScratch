## Create Dataset

```python
## from scratch
import numpy as np
x_1 = np.array([[10], [9], [6], [4]])
x_2 = np.array([[5], [7], [8], [9]])
Y = np.array([ 4,  6,  9, 10 ])
```

## Linear Regression Algorithm

```python
# mean of our inputs and outputs
x1_mean = np.mean(x_1); x2_mean = np.mean(x_2)
y_mean = np.mean(Y)
```

```python
#total number of values
n = len(x_1)
```

```python
# using the formula to calculate the b1 and b0
x1,x2,y,x11,x22,x12,x1y,x2y = 0,0,0,0,0,0,0,0

for i in range(n):
  y += Y[i]
  x1 += x_1[i]
  x2 += x_2[i]
  x11 += x_1[i]**2
  x22 += x_2[i]**2
  x12 += x_1[i]*x_2[i]
  x1y += x_1[i]*Y[i]
  x2y += x_2[i]*Y[i]

b1 = (( (x22 - (x2*x2/n))*(x1y - (x1*y/n)) )-( (x12 - (x1*x2/n)) * (x2y - (x2*y/n)) )) / ( (x11 - (x1*x1/n))*(x22 - (x2*x2/n)) - (x12 - (x1*x2/n))**2    )
b2 = (( (x11 - (x1*x1/n))*(x2y - (x2*y/n)) )-( (x12 - (x1*x2/n)) * (x1y - (x1*y/n)) )) / ( (x11 - (x1*x1/n))*(x22 - (x2*x2/n)) - (x12 - (x1*x2/n))**2    )
b0 = y_mean - (b1 * x1_mean) - (b2 * x2_mean) #printing the coefficient

print(f'coefficient: ( b1:{b1}, b2:{b2}) \nintercept (b0:{b0})')
```
