# KNN algorithm

First, we create a function to calculate the distance with 2 criteria; euclidean and mahhathan

```python
def euclid(y,x):
  return np.sqrt(sum((y - x)**2))

def mahhat(y,x):
  return sum(abs(y-x))


def distance(y,x,method='euclidean'):
  if method == 'euclidean':
    return euclid(y,x)
  
  elif method == 'mahhathan':
    return mahhat(y,x)
  
  else:
    print("We doesn't support that method yet")
    return None
```

```python
knn_test = train_data_x.copy()

knn_test['euclidean distance'] = np.nan
knn_test['mahhathan distance'] = np.nan
```

Here's our KNN with rank

```python
mahhat_test_data = test_data.copy()
euclid_test_data = test_data.copy()

mahhat_test_data['predicted covid'] = np.nan
euclid_test_data['predicted covid'] = np.nan
```

Given k=3 ;

```python
k = 3

#for each test set

for i in range(len(test_data_x)):
  
  #for each training set
  print("\n\n")
  print(f"Test set {i}")
  knn_test['euclidean distance'] = np.nan

  for j in range(len(train_data_x)):
    eu = distance(test_data_x.iloc[i], train_data_x.iloc[j])
    ma = distance(test_data_x.iloc[i], train_data_x.iloc[j],method='mahhathan')
    
    knn_test.loc[j,'euclidean distance'] = eu
    knn_test.loc[j,'mahhathan distance'] = ma

##sort by euclidean 
  print("Using Euclidean distance ")
  knn_test.sort_values(by=['euclidean distance'],axis=0,ascending=True, inplace=True)
  display(knn_test)


  check = train_data.iloc[knn_test.index[0:k]].copy()
  print(f"\n first {k} data based on euclidean distance")

  display(check)

  print(f"result: {check.value_counts('covid').index[0]}\n\n\n")
  euclid_test_data.iloc[i,-1] = check.value_counts('covid').index[0]

  train_data.sort_index(inplace = True) 

##sort by mahhathan 

  print("Using Mahhathan distance ")
  knn_test.sort_values(by=['mahhathan distance'],axis=0,ascending=True, inplace=True)
  display(knn_test)
  check = train_data.iloc[knn_test.index[0:k]].copy()
  print(f"\n first {k} data on mahhathan distance")

  display(check)

  print(f"result: {check.value_counts('covid').index[0]}\n\n-----------\n")
  mahhat_test_data.iloc[i,-1] = check.value_counts('covid').index[0]

  train_data.sort_index(inplace = True)
```
